```python
import cv2
import numpy as np

image_path = '/content/utp-0110-075r.jpg'

img = cv2.imread(image_path)
H, W, _ = img.shape

# Initialize an empty canvas to store the overlayed masks
overlay = np.ones((H, W, 3), dtype=np.uint8)

for result in results:
    for j, mask in enumerate(result.masks.data):
        class_id = result.boxes.cls[j].item()  # Get class ID for each mask
        color = (0, 0, 255) if class_id == 0 else (0, 255, 255)  # Red for class 0, Yellow for class 1

        # Resize the mask to match the dimensions of the canvas
        mask = cv2.resize(mask.cpu().numpy(), (W, H)).astype(np.uint8) * 255

        # Convert single-channel mask to 3-channel mask
        mask = cv2.merge([mask, mask, mask])

        # Apply the color to the mask
        mask = cv2.bitwise_and(mask, color)

        # Overlay the mask on the canvas
        overlay = cv2.addWeighted(overlay, 1, mask, 1, 0)

# Save the overlayed masks as a single image
cv2.imwrite('overlayed_masks.png', overlay)
```

```python
from ultralytics import YOLO
import cv2
import numpy as np

model_path = '/content/runs/segment/train3/weights/best.pt'
image_path = '/content/utp-0110-075r.jpg'

img = cv2.imread(image_path)
H, W, _ = img.shape

model = YOLO(model_path)

results = model(img)

# Initialize masks for each class
decorations_mask = np.zeros((H, W), dtype=np.uint8)
text_area_mask = np.zeros((H, W), dtype=np.uint8)

for result in results:
    for label, box, conf in zip(result.pred[0]['class'], result.pred[0]['xyxy'], result.pred[0]['conf']):
        class_id = int(label)
        if class_id == 0:
            # Generate red pixels for decorations
            x1, y1, x2, y2 = [int(x) for x in box]
            decorations_mask[y1:y2, x1:x2] = 255
        elif class_id == 1:
            # Generate yellow pixels for text_area
            x1, y1, x2, y2 = [int(x) for x in box]
            text_area_mask[y1:y2, x1:x2] = 255

# Save the masks as images
cv2.imwrite('decorations_mask.png', decorations_mask)
cv2.imwrite('text_area_mask.png', text_area_mask)
```
